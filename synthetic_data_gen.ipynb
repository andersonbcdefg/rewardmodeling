```python
from tqdm import tqdm
import openai
from datasets import load_dataset

openai.api_base = "http://localhost:4891/v1"
openai.api_key = "not needed for a local LLM"
```

```python
dolly_prompts = load_dataset("databricks/databricks-dolly-15k", split="train").filter(
    lambda example: example['context'] == ""
)['instruction']
```

```python
model = "mpt-7b-chat"
results = []
for prompt in tqdm(dolly_prompts):
    response = openai.Completion.create(
        model=model,
        prompt=prompt,
        max_tokens=500,
        temperature=0.9,
        top_p=0.95,
        n=2,
        echo=False,
        stream=False
    )
    completion1 = response['choices'][0]['text']
    completion2 = response['choices'][1]['text']
    results.append((prompt, completion1, completion2))
```
